```python
import os
import numpy as np
import pandas as pd
import itertools
import datetime as dt
```

```python
#Change base path before executing code
Base_path = "/home/tennistetris/Documents/Data Analytics/Hall/Inter_hall"
folder = 'Sharpe_quarterly_padded'
Path_to_data = os.path.join(Base_path,folder)
os.chdir(Path_to_data)
files = os.listdir(Path_to_data)
```

```python
Files_to_remove = list('padded_sharpe_data.csv')
#Put the file that are not to be included in analysis in this list
```

```python
db_final = pd.DataFrame(columns = ['Quarter','Sharpe','ID'])
files = list(set(files) - set(Files_to_remove))
for file in files:
    db_temp = pd.read_csv(file,names = ['Quarter','Sharpe'])
    db_temp['ID'] = file.split('.')[0]
    print ('file : ',file)
    db_final = pd.concat([db_final,db_temp])
db_final.to_csv('padded_sharpe_data.csv')
```
