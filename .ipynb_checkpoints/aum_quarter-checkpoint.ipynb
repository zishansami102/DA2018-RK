```python
aum_df.iloc[100495]
x= list(aum_df[aum_df['Timeline Quarter'] == 0].id)
```

```python
aum_df["Timeline Quarter"].unique()
```

```python
y=set(x)
y= list(y)
y
```

```python
for i in y:
    aum_df= aum_df[aum_df['id'] != i]
```

```python
quarter=[]
ind=[]
for i in range(len(aum_df)):
    strin= str(aum_df["Timeline Quarter"].iloc[i])
    month= strin.split()[0]
    year= strin.split()[3]
    date= month+" "+year
    #print(i)
    if(year == '2012'):
        ind.append(i)
        continue
    quarter.append(datetime.strptime(date, "%B %Y").strftime("%Y-%m"))
```

```python
aum_df.drop(aum_df.index[ind], inplace= True)
quarter[:10]
```

```python
print(len(aum_df))
len(quarter)
```

```python
#aum_df['quarter']=""
aum_df["quarter"]= quarter
```

```python
aum_df.drop("Timeline Quarter", axis=1, inplace=True)
aum_df.head()
```
